```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
iris=datasets.load_iris()
```

```python
list(iris.keys())
```

```python
print(iris.DESCR)
```

```python
=iris['data'][:,3:]#petal width
print(X)
```

```python
#是的为1，不是为0
y=(iris['target']==2).astype(np.int) #Iris-virginica类的target值为2
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
log_res=LogisticRegression()
log_res.fit(X,y)
```

```python
X_new=np.linspace(0,3,1000).reshape(-1,1)
y_proba=log_res.predict_proba(X_new)
```

```python
y_proba
```

```python
#只要熟悉就行，可作为绘图模板
plt.figure(figsize=(10,5))#宽、长
decision_boundary=X_new[y_proba[:,1]>=0.5][0]#决策边界
plt.plot([decision_boundary,decision_boundary],[-1,2],'k:',linewidth=2)#绘制决策边界
plt.plot(X_new,y_proba[:,1],'g-',label='Iris-Virginica')
plt.plot(X_new,y_proba[:,0],'b--',label='Not Iris-Virginica')
plt.arrow(decision_boundary,0.08,-0.3,0,head_width=0.05,head_length=0.1,fc='b',ec='b')#画箭头
plt.arrow(decision_boundary,0.92,0.3,0,head_width=0.05,head_length=0.1,fc='g',ec='g')
plt.text(decision_boundary+0.02,0.15,'Decision Boundary',fontsize=16,color='k',ha='center')
plt.xlabel('Peta Width(cm)',fontsize=16)
plt.ylabel('y_proba',fontsize=16)
plt.axis([0,3,-0.02,1.02])
plt.legend(loc='center left',fontsize=16)
```

```python
X=iris['data'][:,(2,3)]
y=(iris['target']==2).astype(np.int)
```

```python
from sklearn.linear_model import LogisticRegression
log_res=LogisticRegression(C=10000)#C是正则化强度的倒数
log_res.fit(X,y)
```

决策边界的绘制：
1.构建坐标数据，合理的范围中，根据实际训练时输入数据来决定；
2.整合坐标点，得到所有测试输入数据坐标点；
3.预测，得到所有点的概率值；
4.绘制等高线，完成决策边界；

```python
 # np.linspace则是将2.9到7等分500份，reshape(-1,1)代表行数根据实际情况，列数为1
x0,x1=np.meshgrid(np.linspace(2.9,7,500).reshape(-1,1),np.linspace(0.8,2.7,200).reshape(-1,1))
```

```python
X_new=np.c_[x0.ravel(),x1.ravel()]
X_new
```

```python
X_new.shape
```

```python
# 回归的predic只是返回预测值（返回所有分类中最大的那个），predict_proba则是返回所有类别的预测值
y_proba=log_res.predict_proba(X_new)
```

```python
x0.shape
```

```python
x1.shape
```

```python
plt.figure(figsize=(10,4))
#y值是0（0代表某个分类）的对应X值的第一个特征值，第二个特征值
plt.plot(X[y==0,0],X[y==0,1],'bs')
plt.plot(X[y==1,0],X[y==1,1],'g^')

zz=y_proba[:,1].reshape(x0.shape)
# contour的意思是等高线
contour=plt.contour(x0,x1,zz,cmap=plt.cm.brg)
plt.clabel(contour,inline=1)
plt.axis([2.9,7,0.8,2.7])
plt.text(3.5,1.5,'NOT Vir',fontsize=16,color='b')
plt.text(6.5,2.3,'Vir',fontsize=16,color='g')
```

```python
X=iris['data'][:,(2,3)]
y=iris['target']
```

```python
softmax_reg=LogisticRegression(multi_class='multinomial',solver='lbfgs')
softmax_reg.fit(X,y)
```

```python
softmax_reg.predict([[5,2]])
```

```python
softmax_reg.predict_proba([[5,2]])
```

```python
x0,x1=np.meshgrid(
    np.linspace(0,8,500).reshape(-1,1),
    np.linspace(0,3.5,200).reshape(-1,1))
X_new=np.c_[x0.ravel(),x1.ravel()]

y_proba=softmax_reg.predict_proba(X_new)
y_predict=softmax_reg.predict(X_new)

zzl=y_proba[:,1].reshape(x0.shape)
zz=y_predict.reshape(x0.shape)

plt.figure(figsize=(10,4))
plt.plot(X[y==2,0],X[y==2,1],'g^',label="Iris-Virginica")
plt.plot(X[y==1,0],X[y==1,1],'bs',label="Iris-Versicolor")
plt.plot(X[y==0,0],X[y==0,1],'yo',label="Iris-Setosa")

 #绘制底面颜色
from matplotlib.colors import ListedColormap
custom_cmap=ListedColormap(['#fafab0','#9898ff','#a0faa0'])
plt.contourf(x0,x1,zz,cmap=custom_cmap)

contour=plt.contour(x0,x1,zzl,cmap=plt.cm.brg)
plt.clabel(contour,inline=1,fontsize=12)
plt.xlabel("Petal length",fontsize=14)
plt.ylabel("petal width",fontsize=14)
plt.legend(loc="center left",fontsize=14)
plt.axis=([0,7,0,3.5])
plt.show()
```
